### Plots - Distribution of Neuroticim Scores in Main & Replication Sample with Updated X-axis Range

*This script was used to plot the distribution of neuroticism scores in the Main Study for the main and the replication sample with an updated x-axis range. It was also used to test these distributions for their normality with a Shapiro-Wilk test.*

### 1. Main Sample

#### 1.1. Import Neuroticism Scores and Plot Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# import data
file_path = r"E:\Data Project 4\Behavioral Data HCP\HCP_behavioral_movie_main_89.csv"
df = pd.read_csv(file_path)

# Histogramm of variable 'NEOFAC_N'
plt.figure(figsize=(8, 6))
plt.hist(df['NEOFAC_N'].dropna(), bins=20, edgecolor='black', color='grey')

# Title and axis labels
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of NEOFAC_N Main Sample (N = 89)')

# Set axis limits
plt.xlim([0, 46])
plt.ylim([0, 15])

# Save plot
plt.savefig("Distribution_N_Scores_Main_updated_axis.png", dpi=300, bbox_inches="tight")

# Show plot
plt.show()
```

#### 1.2. Get Characteristics

```python
min_N = df['NEOFAC_N'].min()
max_N = df['NEOFAC_N'].max()
mean_N = df['NEOFAC_N'].mean()
std_N = df['NEOFAC_N'].std()

print(min_N)
print(max_N)
print(mean_N)
print(std_N)
```

#### 1.3. Test Distribution for Normality

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, probplot

# Shapiro-Wilk Test
neuroticism_scores = df['NEOFAC_N']
stat, p = shapiro(neuroticism_scores)
print(f"Shapiro-Wilk Test statistic = {stat:.4f}, p-value = {p:.4f}")

if p > 0.05:
    print("Data appears to be normally distributed.")
else:
    print("Data does not appear to be normally distributed.")

# Visualizations
sns.histplot(neuroticism_scores, kde=True)
plt.title('Histogram with KDE')
plt.show()

probplot(neuroticism_scores, dist="norm", plot=plt)
plt.title("Q-Q Plot")
plt.show()
```
