### Random Distribution of IS-RSA Values for Paper Figure (Study Procedure)

*This script was used to plot a random distribution of IS-RSA values for the figure in the paper that illustrates the study procedure*

```python
import numpy as np
import matplotlib.pyplot as plt
import os

filepath = r'C:\Users\jop86ib\Documents\Paper 4\Outline\Figures_created_for_outline'
filename = "histogram_random_1"

save_fig_path = os.path.join(filepath,filename)

# generate random scores
#np.random.seed(42)  # for reproducibility
values = np.random.normal(loc=0.0, scale=0.05, size=200)

# scale relasitic (-.2 to .2)
values = np.clip(values, -0.2, 0.2)

# PLot histogram 
plt.figure(figsize=(8, 5))
plt.hist(values, bins=30, edgecolor='black', color='#CC5500', alpha=0.4)
plt.xlabel("IS-RSA correlation coefficient ")
plt.ylabel("Frequency")
plt.grid(False)
plt.ylim(0, 21)

plt.savefig(save_fig_path, dpi=300, bbox_inches='tight')


plt.show()
```
