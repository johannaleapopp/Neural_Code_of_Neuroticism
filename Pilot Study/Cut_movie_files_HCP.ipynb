### 2. Define Function

```python
from moviepy.video.io.VideoFileClip import VideoFileClip

def split_video(input_path, output_folder, clip_times):
    """
    Splits video into smaller clips based on given start and end times in seconds.

    parameter input_path: Path to the input video file (e.g., 'video.mp4').
    parameter output_folder: Folder where the resulting clips will be saved.
    parameter clip_times: List of tuples with start and end times in seconds, 
                       e.g., [(start_time, end_time), ...].
    """
    try:
        # Load the original video
        video = VideoFileClip(input_path)
        
        for idx, (start_time, end_time) in enumerate(clip_times):
            # Ensure times are within the video duration
            if start_time >= video.duration or end_time > video.duration:
                print(f"Scene {idx + 1}: Times are outside the video duration. Skipping.")
                continue
            
            # Create the subclip
            subclip = video.subclipped(start_time, end_time)
            
            # Define the output path and save the clip
            output_path = f"{output_folder}/scene_{idx + 1}.mp4"
            subclip.write_videofile(output_path, codec="libx264", audio_codec="aac")
            print(f"Scene {idx + 1} saved: {output_path}")
        
        video.close()
        print("Done! All scenes have been processed.")
    
    except Exception as e:
        print(f"An error occurred: {e}")

# Example input
# input_video = "input.mp4"
# output_directory = "output_clips"
# clip_times = [
  #  (0, 30),    # Clip from 0 to 30 seconds
   # (75, 120),  # Clip from 75 to 120 seconds
    #(210, 255)  # Clip from 210 to 255 seconds
# ]

# split_video(input_video, output_directory, clip_times)
```
